**Understanding the Pollutants** -

**Carbon Monoxide (CO)**: A colorless, odorless gas primarily generated by incomplete combustion of fuels in vehicles, power plants, and household appliances. CO inhibits oxygen delivery to tissues, leading to headaches, dizziness, and impaired cognitive function at high concentrations. Prolonged exposure can be fatal.

**Ozone (O3)**: A secondary pollutant formed by chemical reactions between nitrogen oxides and volatile organic compounds in the presence of sunlight. Ozone irritates the respiratory system, causing coughing, wheezing, and chest tightness, and exacerbates asthma and chronic obstructive pulmonary disease. Elevated ozone levels also contribute to smog formation.

**Nitrogen Dioxide (NO2)**: A reddish-brown gas primarily emitted from vehicle exhaust and industrial processes. NO2 directly irritates the lungs and contributes to the formation of ground-level ozone and particulate matter, further impacting air quality and human health.

**Particulate Matter 2.5 (PM 2.5)**: Microscopic particles with a diameter of less than 2.5 micrometers, easily inhaled deep into the lungs. PM 2.5 originates from various sources, including vehicle emissions, industrial activities, and dust. Exposure to PM 2.5 is linked to cardiovascular and respiratory diseases, including asthma, bronchitis, and lung cancer.

**Understanding the AQI Categories** -

The Air Quality Index (AQI) helps us understand how clean the air is with simple categories and colors.

**Good (Green, 0-50)**: Like taking a deep breath in a forest! Air quality is excellent, posing minimal health risks. Enjoy outdoor activities!

**Moderate (Yellow, 51-100)**: Air quality is acceptable, though sensitive individuals might experience minor irritation. Consider limiting strenuous outdoor activities if you have concerns.

**Unhealthy for Sensitive Groups (Orange, 101-150)**: Sensitive individuals, like children and those with asthma, may experience breathing difficulties. Consider reducing outdoor activities, especially for these groups.

**Unhealthy (Red, 151-200)**: Everyone can start feeling the effects of air pollution. Reduce outdoor activities and consider wearing a mask, especially if you have health concerns.

**Very Unhealthy (Purple, 201-300)**: Air quality is hazardous! Minimize outdoor activities and stay indoors as much as possible. Vulnerable individuals should take extra precautions.

**Hazardous (Maroon, 301+)**: Emergency conditions! Stay indoors and avoid all outdoor activities. Follow official health advisories to protect your health.

#### 3.1. Univariate Analysis <a class='anchor'></a>

```python
#Inspecting the column names before analysis
df.columns
```

```python
#Describing the dataset for the statistics
df.describe()
```

1. After the Data cleaning, total number of rows we are left with are 13956 rows.
2. Data reveals that the global average AQI value stands at approximately 63.21. This falls within the category of Moderate air quality on the official AQI scale, generally considered acceptable for most healthy people, but it's important to note that significant regional variations exist.

```python
#Plotting the histogram of AQI Value for a better understanding of the distibution
#Chose plotly express for this because of its ability to associate the category with its universally accepted color coding also for its interactive features

histdata = df[["AQI Value", "AQI Category"]]

fig = px.histogram(
    histdata,
    x="AQI Value",
    color="AQI Category",
    barmode="stack",
    title="Distribution of AQI Values",
    color_discrete_sequence=["yellow", "green", "red", "purple","orange", "black"]
)

fig.update_layout(
    xaxis_title="AQI Value",
    yaxis_title="Frequency",
    legend_title="AQI Category",
)

fig.show()
```

```python
#Plotting the histogram of AQI Category for a better understanding of its distibution
color_sequence = ["yellow", "green", "red", "purple","orange", "black"]
sns.countplot(x = "AQI Category", data = df, palette = color_sequence)
plt.xlabel("AQI Category")
plt.ylabel("Frequency")
plt.title("AQI Category Distribution")
plt.xticks(rotation = 40)
plt.show()
```

Based on the analysis of dataset, the distribution of AQI values shows a pronounced concentration within the good and moderate ranges, implying generally acceptable air quality for most healthy people.

```python
# Calculating Ten highest average AQI values for countries and visualising it with a bar chart
avg_aqi_country = df.groupby("Country")["AQI Value"].mean().round(2).reset_index()
topten_polluted_countries = avg_aqi_country.sort_values("AQI Value", ascending = False).head(10)
print(topten_polluted_countries)
fig1 = topten_polluted_countries.plot.bar(x = "Country", title = "Top Ten Countries with Highest average AQI Value")
```

1. Among the countries analyzed in this dataset, the Republic of Korea stands out with one of the highest average AQI values.
2. Data reveals that a significant portion of the countries with high AQI values listed are from the Middle East. This region, characterized by extensive deserts, is often impacted by dust storms, which have been linked to air quality challenges.

```python
# Calculating Ten lowest average AQI values for countries and visualising it with a bar chart
topten_clean_countries = avg_aqi_country.sort_values("AQI Value", ascending = True).head(10)
print(topten_clean_countries)
fig2 = topten_clean_countries.plot.bar(x = "Country", y = "AQI Value", title = "Top Ten Countries with Lowest average AQI Value")
```

1. The island nation of Palau ranks among the countries with the lowest average AQI values in this dataset, suggesting relatively clean air quality.
2. A notable proportion of countries with low AQI values on this list are small island nations with limited industrial activity and a focus on tourism. This suggests a potential link between such economic characteristics and healthier air quality.

```python
# Calculating Ten Highest average AQI values for Cities and visualising it with a bar chart
avg_aqi_city = df.groupby("City")["AQI Value"].mean().round(2).reset_index()
topten_polluted_cities = avg_aqi_city.sort_values("AQI Value", ascending = False).head(10)
print(topten_polluted_cities)
fig3 = topten_polluted_cities.plot.bar(x = "City", title = "Top Ten Cities with Highest average AQI Value")
```

Common factor between them is that they are all situated in areas with high population densities, industrial activities, and vehicle traffic, which contribute to poor air quality. As a result, they all have high AQI (Air Quality Index) values, indicating unsafe levels of air pollution.

```python
# Calculating Ten Lowest average AQI values for Cities and visualising it with a bar chart
topten_clean_cities = avg_aqi_city.sort_values("AQI Value", ascending = True).head(10)
print(topten_clean_cities)
fig4 = topten_clean_cities.plot.bar(x = "City", title = "Top Ten Cities with Lowest average AQI Value")
```

Despite their differences in location and climate, they share a common trait of having low AQI values due to their relatively low levels of industrial activity, vehicle emissions, and population density. As a result, they tend to have cleaner air and lower levels of air pollution compared to major urban centers.

#### 3.2. Bivariate Analysis <a class='anchor'></a>

```python
#Visualising the AQI values of the Cities based on their location with plotly expresses's scatter mapbox
fig = px.scatter_mapbox(df,
                        lon = df['lng'],
                        lat = df['lat'],
                        hover_name = "AQI Value",
                        zoom = 1,
                        color = df['AQI Category'],
                        text= df['City'],
                        width = 1200,
                        height = 800,
                        title = 'World AQI',
                        color_discrete_sequence=["yellow", "green", "red", "purple","orange", "black"]
                        )

fig.update_layout(mapbox_style = "open-street-map")
fig.update_layout(legend=dict(orientation="h",))
fig.update_layout(margin = {"r":0, "t":50, "l":0, "b":10})
fig.show()
```

1. India and China have the most spots with high AQI values due to a combination of factors, including rapid industrialization, large populations, and extensive agricultural burning.
2. North America, Europe, South America generally good air quality, with most regions falling under the "Good" AQI category, can be attributed to a combination of factors, including strict environmental regulations, widespread use of clean energy sources, and reduced industrial activity compared to other regions. 
3. Africa mostly consists of Moderate AQI levels apart from some cities on the East with high population density.

```python
#Select relevant air pollution columns for correlation analysis
air_pollutants = df[['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']]
#Calculate the correlation matrix for the selected pollutants
aqi_corr = air_pollutants.corr()
```

```python
#Create a heatmap to visualize the correlation matrix
sns.heatmap(aqi_corr.round(2), annot = True, cmap ="RdGy")
plt.xlabel("Pollutants")
plt.ylabel("Pollutants")
```

#### 3.3. Multivariate Analysis <a class='anchor'></a>

```python
#Create a pairwise plot to visualize relationships between pollutants
sns.pairplot(air_pollutants)
```

Data reveals that PM2.5 AQI is the most substantial contributor to the AQI value. CO and Ozone AQI values also have prominent influences, followed by NO2 AQI.
